### Pré-processamento do Dataset

```python
# Alterando o nome das variáveis para facilitar a identificação das mesmas.
def renomear_variaveis(dados):
    # Renomeando as variáveis no DataFrame.
    dados.rename(columns={
        'ID Veiculo': 'ID_Veiculo',
        'Ano Fabricacao Veiculo': 'Ano_Fabricacao_Veiculo',
        'Tipo Combustivel': 'Tipo_Combustivel',
        'Capacidade de Carga (Toneladas)': 'Capacidade_Carga',
        'ID Rota': 'ID_Rota',
        'Distancia (KM)': 'Distancia',
        'Uso de Combustivel (Litros)': 'Uso_Combustivel'
    }, inplace=True)
    
    return dados

# Chamando a função para renomear as variáveis.
dados = renomear_variaveis(dados)

# Agora, as variáveis foram renomeadas no DataFrame.
dados.head(15)
```

### Pergunta 2 - Existe alguma correlação entre a distância da rota e o consumo de combustível?

Para determinar se existe alguma correlação entre a distância da rota e o consumo de combustível, podemos calcular o coeficiente de correlação de Pearson entre essas duas variáveis. O coeficiente de correlação de Pearson varia de -1 a 1, onde:

1 indica uma correlação positiva perfeita: à medida que uma variável aumenta, a outra também aumenta na mesma proporção.

-1 indica uma correlação negativa perfeita: à medida que uma variável aumenta, a outra diminui na mesma proporção.

0 indica ausência de correlação linear entre as variáveis.

```python
# Calculando o coeficiente de correlação de Pearson entre a distância e o consumo de combustível.
correlacao = dados['Distancia'].corr(dados['Km_L'])

# Exibindo o coeficiente de correlação.
print(f"Coeficiente de correlação entre distância e consumo de combustível: {correlacao}")
```

```python
# Criando um gráfico de dispersão com linha de tendência.
plt.figure(figsize=(10, 6))
sns.regplot(x='Distancia', y='Km_L', data=dados, scatter_kws={'s': 50}, line_kws={'color': 'red'})
plt.xlabel('Distância da Rota (KM)')
plt.ylabel('Uso de Combustível (Litros)')
plt.title('Correlação entre Distância da Rota e Uso de Combustível')
plt.show()
```

### Resposta: Como o Coeficiente de correlação entre distância e consumo de combustível(Km/L) foi de 0.7975961487690542, podemos afirmar que existe uma alta correlação positiva entre estas 2 variáveis.

Verificado que existem outliers na Km_L, mas vamos validar esta informação.

```python
def verificar_outliers(dataframe, variavel):
    # Calculando o primeiro e terceiro quartis.
    Q1 = dataframe[variavel].quantile(0.25)
    Q3 = dataframe[variavel].quantile(0.75)
    
    # Calculando o IQR (Intervalo Interquartil).
    IQR = Q3 - Q1
    
    # Definindo os limites para identificar outliers.
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    
    # Verificando os outliers.
    outliers = dataframe[(dataframe[variavel] < limite_inferior) | (dataframe[variavel] > limite_superior)]
    
    return outliers

# Verificar outliers para a variável 'Uso_Combustivel'.
outliers_uso_combustivel = verificar_outliers(novo_dataset, 'Km_L')
print("Outliers em 'Km_L':")
print(outliers_uso_combustivel)

# Verificar outliers para a variável 'ID_Rota'.
outliers_id_rota = verificar_outliers(novo_dataset, 'ID_Rota')
print("\nOutliers em 'ID_Rota':")
print(outliers_id_rota)

# Verificar outliers para a variável 'Distancia'.
outliers_distancia = verificar_outliers(novo_dataset, 'Distancia')
print("\nOutliers em 'Distancia':")
print(outliers_distancia)
```

## Pre-Processamento dos Dados

```python
# Função para remover outlier da coluna Km_L.
def remover_outliers(dataframe, coluna):
    # Calculando o primeiro e terceiro quartis.
    Q1 = dataframe[coluna].quantile(0.25)
    Q3 = dataframe[coluna].quantile(0.75)
    
    # Calculando o IQR (Intervalo Interquartil).
    IQR = Q3 - Q1
    
    # Definindo os limites para identificar outliers.
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    
    # Removendo outliers.
    dataframe_sem_outliers = dataframe[(dataframe[coluna] >= limite_inferior) & (dataframe[coluna] <= limite_superior)]
    
    return dataframe_sem_outliers

# Remover outliers da coluna 'Uso_Combustivel'.
novo_dataset1 = remover_outliers(novo_dataset, 'Km_L')

# Verificar o número de observações antes e depois da remoção de outliers.
print("Número de observações antes da remoção de outliers:", len(novo_dataset))
print("Número de observações após a remoção de outliers:", len(novo_dataset1))
```

```python
# Criando subplots para os histogramas após a remoção dos outliers.
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))
fig.subplots_adjust(wspace=0.5)

# Criando histogramas para cada variável.
for i, variavel in enumerate(variaveis):
    sns.histplot(novo_dataset[variavel], bins=30, ax=axs[i], kde=True)
    axs[i].set_title(f'Histograma - {variavel}')

plt.show()
```

###  Construção, Treinamento e Avaliação do Modelo 3 com SVM com grid search

```python
# Define a grade de hiperparâmetros para pesquisar.
param_grid = {'C': [0.1, 1, 10, 100],
              'kernel': ['linear', 'rbf', 'poly'],
              'gamma': [0.1, 1, 10]}
```

```python
# Cria o objeto GridSearchCV.
grid_search = GridSearchCV(SVR(), param_grid, cv=5)
```

```python
# Realize a pesquisa em grade nos dados de treino.
grid_search.fit(X_treino, y_treino)
```

```python
# Obtém os melhores hiperparâmetros encontrados pelo Grid Search.
melhores_hiperparametros = grid_search.best_params_
```
